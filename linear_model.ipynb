```python
import eli5
import pandas as pd
import sklearn
```

```python
df = pd.read_csv("iris.csv")
df
```

```python
from sklearn.preprocessing import LabelEncoder

df["target"] = LabelEncoder().fit_transform(df.variety)
df.head()
```

```python
X = df.drop(["variety", "target"], axis=1)
y = df.target

model = sklearn.linear_model.LogisticRegression(max_iter=1000)
model.fit(X, y)
```

```python
eli5.show_weights(model)
```

```python
eli5.explain_weights(model)
```

```python
eli5.explain_prediction(model, X.loc[0])
```
